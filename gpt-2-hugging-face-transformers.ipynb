# GPT-2: Hugging Face Transformers
---
This is how you can simply use the `transformers` library from Hugging Face to generate answer from the pretrained model GPT-2.

# Imports
We need to import the pretrained model and the tokenizer.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

# Load the GPT-2 model

```python
model_name = 'gpt2-medium'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

# Input text and generate answer

```python
input_text = "Hello, how are you?"

input_ids = tokenizer.encode_plus(input_text, add_special_tokens=True, return_tensors='pt')
attention_mask = input_ids['attention_mask']
input_ids = input_ids['input_ids']

generated_text = model.generate(input_ids=input_ids,
                                attention_mask=attention_mask,
                                do_sample=True)
print(tokenizer.decode(generated_text[0], skip_special_tokens=True))
```
